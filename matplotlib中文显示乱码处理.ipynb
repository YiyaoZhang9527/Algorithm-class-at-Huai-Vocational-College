## 查看当前环境有哪些字体可以显示中文

```python
import warnings
warnings.filterwarnings('ignore')
from matplotlib.font_manager import FontManager
fm = FontManager()
mat_fonts = set(f.name for f in fm.ttflist)
n = 0
for mat_font in mat_fonts:
    n += 1
    print(f"当前序号:[{n}]<->当前字体:[{mat_font}]")
    plt.rcParams['font.sans-serif'] = [mat_font] # mac os 中文字体
    plt.rcParams['axes.unicode_minus'] = False
    plt.figure(figsize=(2,2))
    plt.title("-中文测试是否能用")
    plt.xlabel('x轴数据')
    
    plt.ylabel('y轴数据')
    plt.show()
```

```python
#!/usr/bin/env python
# -*- coding:utf-8 -*-
 
# 定积分函数，可以修改
def fx(x):
    return -x ** 2 + 2 * x
 
# 定义第i个区间的中点值，即定义积分变量
def middle(a, b, n, i):
    return a + i * (b - a) / n;
 
# 定义每个小区间的间隔差，即将范围分成n个等区间
def cha(a, b, n):
    return (b - a) / n;
 
# 函数体
def fun():
    # 抛物线与x轴相交于a(0, 0), b(0, 2)
    a = 0
    b = 2
    # 执行次数，次数越大，结果越近真实数值
    n = 100000
    sum = 0
    e = cha(a, b, n)
    for i in range(20, n):
        x = middle(a, b, n, i)
        sum += fx(x)
    print("面积是：%f" % (sum * e))

if __name__ == '__main__':
    fun()
```
