## 查看当前环境有哪些字体可以显示中文

```python
import warnings
warnings.filterwarnings('ignore')
from matplotlib.font_manager import FontManager
fm = FontManager()
mat_fonts = set(f.name for f in fm.ttflist)
n = 0
for mat_font in mat_fonts:
    n += 1
    print(f"当前序号:[{n}]<->当前字体:[{mat_font}]")
    plt.rcParams['font.sans-serif'] = [mat_font] # mac os 中文字体
    plt.rcParams['axes.unicode_minus'] = False
    plt.figure(figsize=(1,1))
    plt.title("-中文测试是否能用")
    plt.xlabel('x轴数据')
    
    plt.ylabel('y轴数据')
    plt.show()
```
