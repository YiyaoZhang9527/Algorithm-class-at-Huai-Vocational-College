```python
import numpy as np
X =np.array([
    [1,1]
    ,[2,1]
    ,[3,1.2]
    ,[4,1]
    ,[5,0.8]
    ,[6,1]
])
X
```

```python
from sklearn.decomposition import NMF

model = NMF(n_components=2,init='random',random_state=0)
W = model.fit_transform(X)
W
```

```python
H = model.components_
H
```

```python
W.dot(H)
```

#### 图像分解

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("./data/old.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 

print(img.shape)
plt.imshow(img)
plt.show()
```

```python
from sklearn.decomposition import NMF

model = NMF(n_components=80,init='random',random_state=12)
W = model.fit_transform(img)
# plt.imshow(W)
# plt.show()
```

```python
W.shape
```

```python
H = model.components_
H.shape
```

```python
plt.imshow(W.dot(H))
```
