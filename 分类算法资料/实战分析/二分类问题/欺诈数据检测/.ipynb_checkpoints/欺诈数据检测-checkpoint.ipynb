数据 Data

📂 Transaction Data 交易数据

transaction_records.csv：包含交易记录以及交易 ID、日期、金额和客户 ID 等详细信息。

transaction_metadata.csv：包含每个事务的附加元数据。

📂 Customer Profiles 客户档案

customer_data.csv：包含客户资料，其中包含姓名、年龄、地址和联系方式等信息。

account_activity.csv：提供客户帐户活动的详细信息，包括帐户余额、交易历史记录和帐户状态。

📂 Fraudulent Patterns 欺诈模式

fraguise_indicators.csv：包含欺诈模式和可疑活动的指标。

suspicious_activity.csv.csv：提供标记为可疑的交易的具体详细信息。

📂 Transaction Amounts 交易金额

amount_data.csv：包括每笔交易的交易金额。

anomaly_scores.csv：提供交易金额的异常分数，表明潜在的欺诈行为。

📂 Merchant Information 商户信息

商户数据.csv：包含参与交易的商户的信息。

transaction_category_labels.csv：提供不同交易类型的类别标签。

## 数据导入检查
#### 1.0 列出csv数据文件的地址

```python
def process_directory(dirpath):
    all_files,all_dirs=[],[]
    for root, dirs, files in os.walk(dirpath):
        for file in files:
            all_files.append(os.path.join(root,file))
        for dir in dirs:
            all_dirs.append(os.path.join(root,dir))
    return all_files,all_dirs

csvfiles_path = process_directory(dirpath)[0]
csvfiles_path
```

#### 1.4 合并数据

```python
combining_tables = pd.merge(transaction_metadata,transaction_records,on="TransactionID")
combining_tables = pd.merge(combining_tables,amount_data,on="TransactionID")
combining_tables = pd.merge(combining_tables,anomaly_scores,on="TransactionID")
combining_tables = pd.merge(combining_tables,transaction_category_labels,on="TransactionID")
combining_tables = pd.merge(combining_tables,fraud_indicators,on="TransactionID")
combining_tables = pd.merge(combining_tables,merchant_data,on = "MerchantID")
combining_tables = pd.merge(combining_tables,suspicious_activity,on="CustomerID")
combining_tables = pd.merge(combining_tables,account_activity,on="CustomerID")
combining_tables = pd.merge(combining_tables,customer_data,on="CustomerID")
combining_tables
```

```python
combining_tables.rename(columns={"TransactionID":"交易ID"
                                 ,"Timestamp":"时间戳"
                                 ,"MerchantID":"商家ID"
                                 ,"Amount":"金额"
                                 ,"CustomerID":"客户ID"
                                 ,"TransactionAmount":"交易金额"
                                 ,"AnomalyScore":"异常分数"
                                 ,"Category":"消费类别"
                                 ,"FraudIndicator":"欺诈指标"
                                 ,"MerchantName":"商家名称"
                                 ,"Location":"地点"
                                 ,"SuspiciousFlag":"可疑标志"
                                 ,"AccountBalance":"帐户余额"
                                 ,"LastLogin":"最后登录时间"
                                 ,"Name":"姓名"
                                 ,"Age":"年龄"
                                 ,"Address":"地址"}
                                 ,inplace=True)
combining_tables
```

```python
from datetime import datetime,timedelta
import time

import pandas as pd
# 构造时间数据
date_time_str_list = [
    '2019-01-01 01:22:26', '2019-02-02 04:34:52', '2019-03-03 06:16:40',
    '2019-04-04 08:11:38', '2019-05-05 10:52:39', '2019-06-06 12:06:25',
    '2019-07-07 14:05:25', '2019-08-08 16:51:33', '2019-09-09 18:28:28',
    '2019-10-10 20:55:12', '2019-11-11 22:55:12', '2019-12-12 00:55:12',
]
df = pd.DataFrame({'时间': date_time_str_list})
# 把字符串格式的时间转换成Timestamp格式
df['时间'] = df['时间'].apply(lambda x: pd.Timestamp(x))

# 年份
df['年']=df['时间'].apply(lambda x: x.year)

# 月份
df['月']=df['时间'].apply(lambda x: x.month)

# 日
df['日']=df['时间'].apply(lambda x: x.day)

# 小时
df['时']=df['时间'].apply(lambda x: x.hour)

# 分钟
df['分']=df['时间'].apply(lambda x: x.minute)

# 秒数
df['秒']=df['时间'].apply(lambda x: x.second)

# 一天中的第几分钟
df['一天中的第几分钟']=df['时间'].apply(lambda x: x.minute + x.hour*60)

# 星期几；
df['星期几']=df['时间'].apply(lambda x: x.dayofweek)

# 一年中的第几天
df['一年中的第几天']=df['时间'].apply(lambda x: x.dayofyear)

# 一年中的第几周
df['一年中的第几周']=df['时间'].apply(lambda x: x.week)

# 一天中哪个时间段：凌晨、早晨、上午、中午、下午、傍晚、晚上、深夜；
period_dict ={
    23: '深夜', 0: '深夜', 1: '深夜',
    2: '凌晨', 3: '凌晨', 4: '凌晨',
    5: '早晨', 6: '早晨', 7: '早晨',
    8: '上午', 9: '上午', 10: '上午', 11: '上午',
    12: '中午', 13: '中午',
    14: '下午', 15: '下午', 16: '下午', 17: '下午',
    18: '傍晚',
    19: '晚上', 20: '晚上', 21: '晚上', 22: '晚上',
}
df['时间段']=df['时'].map(period_dict)

# 一年中的哪个季度
season_dict = {
    1: '春季', 2: '春季', 3: '春季',
    4: '夏季', 5: '夏季', 6: '夏季',
    7: '秋季', 8: '秋季', 9: '秋季',
    10: '冬季', 11: '冬季', 12: '冬季',
}
df['季节']=df['月'].map(season_dict)
df
```
