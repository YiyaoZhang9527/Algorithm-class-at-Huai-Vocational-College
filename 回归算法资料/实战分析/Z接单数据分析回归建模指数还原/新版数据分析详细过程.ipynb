## 统计函数实现
#### 协方差 

```python
def COV(x,y):
    return ((x-np.mean(x))*(y-np.mean(y))).mean()

x = np.arange(100)
COV(x,x)
```

#### 样本方差 

```python
def VAR(x):
    n = x.size
    return np.sum((x-np.mean(x))**2)/(n-1)

VAR(x),np.var(x,ddof=1)
```

#### 总体标准差

```python
def STD(x):
    n = x.size
    var = np.sum((x-np.mean(x))**2)/n
    return np.sqrt(var)

STD(x),np.std(x)
```

#### 皮尔逊系数

```python
def PPMCC(x,y):
    return COV(x,y)/(STD(x)*STD(y))

PPMCC(x,x),pearsonr(x,x)
```
