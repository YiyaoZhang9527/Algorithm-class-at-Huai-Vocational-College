```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.arange(-10,10.0001,0.0001)/0.1
plt.title("x/0.1")
plt.plot(x,c='r')
plt.show()
exp_values = np.exp(x)
plt.title("exp(x/0.1)")
plt.plot(exp_values,c='g')
plt.show()
print(exp_values.max())
```
