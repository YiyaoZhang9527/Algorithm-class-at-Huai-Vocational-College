```python
import fitz  # PyMuPDF库的别名

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)  # 打开PDF文档

    text = ""
    for page_num in range(doc.page_count):
        page = doc[page_num]  # 获取页面
        text += page.get_text()

    doc.close()  # 关闭文档
    return text

# 指定PDF文件路径
pdf_file_path = pdf_paths[0]

# 调用函数并打印结果
extracted_text = extract_text_from_pdf(pdf_file_path)
print(extracted_text)
```

```python
import jieba

jieba.load_userdict("jieba_dict.txt")

def check_keywords_from_extracted_text(keywords,extracted_text):
     
    tokens = list(jieba.cut_for_search(extracted_text))
    word_frequency = {}
    for keyword in keywords:
        
        for token in tokens:
            if keyword == token:
                if keyword in word_frequency:
                    word_frequency[keyword]+=1
                else:
                    word_frequency.update({keyword:1})
    kn = len(word_frequency)
    return kn,word_frequency

keywords = ["电子信息大类","智能","人工智能","网络安全","网络",'python',"网络工程","云计算"]

check_keywords_from_extracted_text(keywords,extracted_text)    
```

### 查看电子信息类有多少

```python
import pandas as pd
import os
import jieba
from tqdm import tqdm

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

keywords = ["电子信息大类","电子与信息大类","人工智能","集成电路","arm","大数据","云计算","openstack"
            ,"docker","数通网络","服务器运维","单片机","网络安全","网络工程",'python',"云计算"
            ,"运维","算法","数据分析"]

kn_list = []
path_list = []
pdf_name_list = []
word_frequency_list = [] 
n = -1
for pdf_file_path in tqdm(pdf_paths,desc="正在对文章信息重要度进行分析"):
    extracted_text = extract_text_from_pdf(pdf_file_path)
    kn,word_frequency = check_keywords_from_extracted_text(keywords,extracted_text)
    if kn > 0:
        n += 1 
        path_list.append(pdf_file_path)
        kn_list.append(kn)
        pdf_name = pdf_file_path.split(os.sep)[-1]
        pdf_name_list.append(f"{n}-{pdf_name}")
        word_frequency_list.append(word_frequency)
        

tables = pd.DataFrame({"关键词数":kn_list,"词频字典":word_frequency_list,"文件名":pdf_name_list,"文件地址":path_list})
tables
```

```python
tables.to_csv("跟竞赛有关建议阅读的pdf.csv")
```

## 将符合条件的pdf移动到指定文件

```python
import shutil

# 指定目标文件夹路径
destination_folder = './需要阅读的pdf/'

# shutil.rmtree(destination_folder)
# os.makedirs(destination_folder)

files_in_folder = os.listdir(destination_folder)

# 循环删除文件夹内的所有文件和子文件夹
for file_or_folder in files_in_folder:
    file_or_folder_path = os.path.join(destination_folder, file_or_folder)
    
    if os.path.isfile(file_or_folder_path):
        os.remove(file_or_folder_path)  # 删除文件
    elif os.path.isdir(file_or_folder_path):
        os.rmdir(file_or_folder_path)  # 删除子文件夹

for pdf_file_path,pdf_name in zip(path_list,pdf_name_list):
    shutil.copy(pdf_file_path,f'{destination_folder}{pdf_name}')
```
