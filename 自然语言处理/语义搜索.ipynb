```python
import os

dir_path = "./2023年安徽省职业院校技能竞赛（高职组）竞赛规程/"
pdf_paths = [f"{dir_path}{file_name}" for file_name in os.listdir(dir_path) if '.pdf' in file_name ]
pdf_paths
```

```python
import fitz  # PyMuPDF库的别名
from tqdm import tqdm
import re

def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)  # 打开PDF文档

    text = ""
    for page_num in range(doc.page_count):
        page = doc[page_num]  # 获取页面
        text += page.get_text()

    doc.close()  # 关闭文档
    return text

def keep_only_chinese_english_space(text):
    # 使用正则表达式只保留中文、英文和空格
    cleaned_text = re.sub(r'[^\u4e00-\u9fa5a-zA-Z\s]', '', text)
    return cleaned_text

def remove_newline(text):
    # 使用正则表达式删除换行符
    cleaned_text = re.sub(r'\n', '', text)
    return cleaned_text

documents = []
for pdf_file_path in tqdm(pdf_paths,desc="正在加载文章"):
    extracted_text = extract_text_from_pdf(pdf_file_path)
    
    documents.append(remove_newline(keep_only_chinese_english_space(extracted_text)))
```

```python
from transformers import BertTokenizer, BertModel
import torch
 
tokenizer = BertTokenizer.from_pretrained("bert-base-chinese")
model = BertModel.from_pretrained("bert-base-chinese")
```

```python
import numpy as np
# Tokenize and encode the documents
document_embeddings = np.zeros((len(documents),768))
document_embeddings
```

```python
n = -1
for document in tqdm(documents,desc='正在对文档进行embeddings'):
    n += 1
    inputs = tokenizer(document, return_tensors="pt", padding=True, truncation=True)
    outputs = model(**inputs)
    document_embedding = outputs.last_hidden_state.mean(dim=1)  # Average over tokens
    document_embedding = document_embedding.detach().numpy()
    # print(document_embedding.shape)
    document_embeddings[n] = document_embedding
# document_embeddings = torch.cat(document_embeddings)
```

### 建立倒排表

```python
import jieba

jieba.load_userdict("jieba_dict.txt")

def jieba_tokenizer(extracted_text):
    return jieba.cut_for_search(extracted_text)



def remove_chinese_stopwords(text):
    # 定义需要删除的助词和语气词
    stopwords = ['的', '了', '着', '是', '在', '和', '有', '很', '就', '不', '都', '一', '一个', '上', '也', '到', '说', '要', '去']

    # 构建正则表达式
    pattern = '|'.join(stopwords)

    # 使用正则表达式删除匹配到的词汇
    cleaned_text = re.sub(pattern, '', text)

    return cleaned_text


def build_inverted_index(documents):
    inverted_index = {}
    
    doc_id = -1
    for document in documents:
        document = remove_chinese_stopwords(document.lower())
        doc_id += 1
        tokens = jieba_tokenizer(document)
        
        for word in tokens:
            if word in inverted_index:
                if doc_id not in inverted_index[word]:
                    inverted_index[word].append(doc_id)
            else:
                inverted_index[word]=[doc_id]
    
    inverted_index.pop(" ")
    return inverted_index
                

inverted_index = build_inverted_index(documents)
```

# 这里输入要搜索的文档内容

```python
user_query = "短视频，创作，运营"
```

```python
user_query_tokens = list(jieba.cut(keep_only_chinese_english_space(remove_chinese_stopwords(user_query.lower()))))
print("搜索关键词:",user_query_tokens)
have_keywords_doc_ids = []
keywords_indexs = {}
for word in user_query_tokens:
    try:
        have_keywords_doc_ids+=inverted_index[word]
    except Exception as e:
        print(f'{e}:{word}',)
    for index in inverted_index[word]:
        if index in keywords_indexs:
            keywords_indexs[index].append(word)
        else:
            keywords_indexs.update({index:[word]})

for ids in set(have_keywords_doc_ids):
    print(f"含有关键词的文档编号:{ids},包含关键词{keywords_indexs[ids]},地址：{pdf_paths[ids]}")

numpy_index = np.zeros(len(documents)).astype(bool)

for doc_id in have_keywords_doc_ids:
    numpy_index[doc_id]=True
```

# 语义近似搜索

```python
user_query_inputs = tokenizer(user_query, return_tensors="pt", padding=True, truncation=True)
user_query_outputs = model(**user_query_inputs)
user_query_embedding = user_query_outputs.last_hidden_state.mean(dim=1)

from sklearn.metrics.pairwise import cosine_similarity
# 计算用户查询与所有文档之间的余弦相似度
similarities = cosine_similarity(user_query_embedding.detach().numpy(), document_embeddings[numpy_index])
# 查找最相似文档的索引
most_similar_document_index = similarities.argmax()
have_keywords_doc_ids[most_similar_document_index]
```

```python
most_similar_document_path = pdf_paths[have_keywords_doc_ids[most_similar_document_index]]
print("最相似的文档:", most_similar_document_path)
for line in extract_text_from_pdf(most_similar_document_path).split('\n'):
    print(line)
```
